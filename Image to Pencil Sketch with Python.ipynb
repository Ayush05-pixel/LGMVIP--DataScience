## Task 4- Image to Pencil Sketch with Python
This is a short project aimed at converting RGB images into pencil sketches. The main library that we are going to use for this purpose is OpenCV.

### Importing Libraries

```python
import cv2
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

### Loading and Plotting the Original Image

```python
img = cv2.imread('Dog.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis('off')
plt.title('Original Image')
plt.show()
```

### Converting the Original Image to Grayscale

```python
img_grayscale = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(8,8))
plt.imshow(img_grayscale, cmap='gray')
plt.axis('off')
plt.title('GrayScale Image')
plt.show()
```

### Inversion of Image

```python
img_invert = cv2.bitwise_not(img_grayscale)
plt.figure(figsize=(8,8))
plt.imshow(img_invert, cmap='gray')
plt.axis('off')
plt.title('Inverted Image')
plt.show()
```

### Making inverted image smoother

```python
img_smooth = cv2.GaussianBlur(img_invert, (21,21), sigmaX=0, sigmaY=0)
plt.figure(figsize=(8,8))
plt.imshow(img_smooth, cmap='gray')
plt.axis('off')
plt.title('Smooth Image')
plt.show()
```

### Converting the Smooth Image to Pencil Sketch

```python
pencil_sketch = cv2.divide(img_grayscale, 255-img_smooth, scale=255)
plt.figure(figsize=(8,8))
plt.imshow(pencil_sketch, cmap='gray')
plt.axis('off')
plt.title('Final Pencil Sketch Image')
plt.show()
```

### Displaying all Images

```python
plt.figure(figsize=(20,20))
plt.subplot(1,5,1)
plt.imshow(img)
plt.axis("off")
plt.title("Original Image")
plt.subplot(1,5,2)
plt.imshow(img_grayscale,cmap="gray")
plt.axis("off")
plt.title("GrayScale Image")
plt.subplot(1,5,3)
plt.imshow(img_invert,cmap="gray")
plt.axis("off")
plt.title("Inverted Image")
plt.subplot(1,5,4)
plt.imshow(img_smooth,cmap="gray")
plt.axis("off")
plt.title("Smooth Image")
plt.subplot(1,5,5)
plt.imshow(pencil_sketch,cmap="gray")
plt.axis("off")
plt.title("Final Pencil Sketch Image")
plt.show()
```
